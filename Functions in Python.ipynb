<b> Doing it Via Function </b>

```python
def buy_books(unbought_books,bought_books):
    while unbought_books:
        currently_at_home = unbought_books.pop()
        print(f"Buying the books: {currently_at_home}")
        bought_books.append(currently_at_home)
    
def show_my_books(bought_books):
    print("\n\nThe following books has been already bought.\n")
    for i in bought_books:
        print(i)

unbought_books = ["A crash Course on Python","Python Data Science Handbook","Introduction to Machine Learning","Mathematics for Machine Learning"]
bought_books = []

buy_books(unbought_books,bought_books)
show_my_books(bought_books)
```

<h4>7. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters. <br>
Sample String : 'The quick Brow Fox' <br>
Expected Output :<br>
No. of Upper case characters : 3<br>
No. of Lower case Characters : 12</h4>

```python
def find_upper_lower(sample):
    upper = 0
    lower = 0
    for i in range(len(sample)):
        if ord(sample[i])>=65 and ord(sample[i])<=90:
            upper +=1
        elif ord(sample[i])>=97 and ord(sample[i])<=122:
            lower +=1
        
    print("No. of Upper case characters :",upper)
    print("No. of Lower case characters :",lower)
find_upper_lower('The quick Brow Fox') 
```

<h3>8. Write a Python function that takes a list and returns a new list with unique elements of the first list. <br>
Sample List : [1,2,3,3,3,3,4,5]<br>
Unique List : [1, 2, 3, 4, 5]<br></h3>

```python
def unique_check(sample_list):
    
    unique_list = []
    for i in sample_list:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

print(unique_check(sample_list = [1,2,3,3,3,4,5]))
```

<h3> Recursion Function </h3>

```python
#recursion means a function can call itself.
"""In the factorial program we saw:
5! = 5*4*3*2*1
4! = 4*3*2*1
3! = 3*2*1
2! = 2*1
1! = 1
n! = n*(n-1)!
 """

def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(5))
```

<h2> Pure Functions </h2>

```python
# Create a global variable `A`.
A = 5

def impure_sum(b):
    # Adds two numbers, but uses the
    # global `A` variable.
    return b + A

def pure_sum(a, b):
    # Adds two numbers, using
    # ONLY the local function inputs.
    return a + b

print(impure_sum(6))
#11

print(pure_sum(4, 6))
#10
```
