# List Vs Array

### Similiarities---
* They Both stores data.
* They both are mutable, we can change (add/delete---) the data inside the list and arrays.
* They can be indexed. 
* They can be sliced. We can accese data in both of them.
* They are ordered.

### Differences---
* A list can contain different kind of data in the same time. like a list can have, string, float, integer etc.<br> ex: list = [1,'rubayet',56.89,'s']<br> But all the data must have same type in array. **The data types must be same for an array**.<br> Ex: array=[1,2,3,4] or [1.0,2.0,3.0,4.0]

* array takes less space in our storage, where list occupy more space.
* array is way more faster than list
* Arrays are great for numerical operations; lists cannot directly handle math operations. For example, you can divide each element of an array by the same number with just one line of code
* List elements doesn't need to be unique, but all the elements inside an array must be unique.

# Array
 An array is also a data structure that stores a collection of items. <br>
 We can create array using- array module, numpy library <br>
 * One Dimensional Array - 1D Array: have only one row or only one column. 
 * ex : [1,2,3,4,5]
 * Two Dimensional Array - 2D Array: have 2 or more than 2 rows/columns.
 * ex: ([[1,2],[3,4]])


# linspace()

* it returns evenly spaced numbers over a specified interval.
* linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)= [by default value]
* when endpoint is true, the stop value is included. when endpoint is false, the stop value is excluded.
* by retstep we can see the difference in each elements. 
* axis = 0, means the array will have less column, all the values are set out column-wise.
* axis = 1, means the array will have more column, all the values are set out row-wise.

```python
help(np.linspace)
```

# ravel()

* flatten always returns a copy.
* ravel returns a view of the original array whenever possible. If you modify the entries in an array returned from flatten this will never happen. ravel will often be faster since no memory is copied, but you have to be more careful about modifying the array it returns

## calculating regression line
* formula: 


```python
df2 = pd.read_csv('ice_cream.csv')
```

```python
ice = df2.to_numpy()
```

```python
tem_mean = np.mean(ice[:,0])
```

```python
sales_mean = np.mean(ice[:,1])
```

```python
numerator = np.sum(((ice[:,0]-tem_mean)* (ice[:,1]-sales_mean)))
```

```python
denominator = np.sum(np.square(ice[:,0]-tem_mean))
```

```python
slope = numerator/denominator
```

```python
y_intercept = sales_mean - slope * tem_mean
```

```python
y_intercept
```

```python
regression_array = ice[:,0] * slope + y_intercept
```

```python
regression_array
```

# Integral and Derivatives
