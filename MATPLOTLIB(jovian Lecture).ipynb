<h2>Styling Lines and Markers</h2>
<br> color or <b> c </b> : b, g, r, c, m, y, k =black, w
<br> linestyle or <b> ls </b> : Solid or dashed Lines
<br> linewidth or <b> lw </b> : set width of a line
<br> markersize or <b> ms </b>: Set the size of markers
<br> markeredgecolor or <b> mec </b> : set the edge color for markers
<br> markeredgewidth or <b> mew </b> : set the edge with markers
<br> markerfacecolor or <b> mfc </b> : set the fill color for markers
<br> alpha : opacity of the plot


```python
plt.plot(years, yield_apples, marker = 's', c= "m", ls = "-", lw = 2, ms = 8, mew =2 , mec = "navy" )
plt.plot(years, oranges, marker = 'o', c= "g", ls = "--", lw = 3, ms = 10, mew =2 , mec = "orange", alpha = 0.5 )
plt.xlabel("Year")
plt.ylabel("Yield(Tons Per Hectare)")
plt.title("Crop Yield in PanchGar")
plt.legend(['Apples','Oranges']);
```

there's short way to write :
<br>fmt  = '[marker][line][color]'
<br>'g--r' = marker shape = o (round), linestyle = --, linecolor = red

```python
plt.plot(years, yield_apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel("Year")
plt.ylabel("Yield(Tons Per Hectare)")
plt.title("Crop Yield in PanchGar")
plt.legend(['Apples','Oranges']);
```

if no <b>linestyle</b> is given then, only markers are drawn in <b>fmt</b> format.

```python
plt.plot(years, oranges, 'or');
plt.xlabel("Year")
plt.ylabel("Yield(Tons Per Hectare)")
plt.title("Crop Yield in PanchGar")
```

<h1>Multiple Histograms</h1>

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins =np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins =np.arange(2,5,0.25));
plt.legend(['setosa','versicolor'])
plt.title('comparing distribution of Sepal Width');
```

when 2 histogram is overlaping, we must use opacity(alpha).

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins = np.arange(2,5,0.25),
         stacked = True)
plt.legend(['Setosa','Versicolor','Virginica'])
```
